```python
%matplotlib inline
import math
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)

all_schedules = []

for p in range(1, 13):
    if p % 2 == 1:
        k = int(16 - (p-1) / 2)
        if p < 10:
            name = "db/" + "0" + str(p) + "_beosztas_20" + str(k) + "_tavasz.xls"
            schedule = pd.read_excel(name , usecols=lambda x: 'Unnamed' not in x)
        else:   
            name = "db/" + str(p) + "_beosztas_20" + str(k) + "_tavasz.xls"
            schedule = pd.read_excel(name, usecols=lambda x: 'Unnamed' not in x)
        schedule['Év'] = 2000 + k;
    else:
        k = int(16 - p / 2)
        if p < 10:
            name = "db/" + "0" + str(p) + "_beosztas_20" + str(k) + "_osz.xls"
            schedule = pd.read_excel(name, usecols=lambda x: 'Unnamed' not in x)
        else:
            name = "db/" + str(p) + "_beosztas_20" + str(k) + "_osz.xls"
            schedule = pd.read_excel(name, usecols=lambda x: 'Unnamed' not in x)
        schedule['Év'] = 2000 + k;
    schedule['Vizsgaztato1'] = schedule['Vizsgáztató'].str.split("\n").str[0]
    schedule['Vizsgaztato2'] = schedule['Vizsgáztató'].str.split("\n").str[1]
    schedule['Vizsgatargy1'] = schedule['Vizsgatárgy'].str.split("\n").str[0]
    schedule['Vizsgatargy2'] = schedule['Vizsgatárgy'].str.split("\n").str[1]
    all_schedules.append(schedule)

data = pd.concat(all_schedules, axis = 0)
data
```

```python
data.dtypes
```

```python
teachers = teachers = [cell.split('\n') for cell in data['Vizsgáztató'] if cell is not np.nan]
teachers = [item for sublist in teachers for item in sublist]
teachers = set(teachers)
teachers
```

```python
studentPerCourse = {}

for i,k in data.iterrows():
    if pd.notnull(k['Vizsgaztato1']):
        if pd.notnull(k['Vizsgatargy1']):
            studentPerCourse[k['Vizsgatargy1']] = 0
        if pd.notnull(k['Vizsgatargy2']):
            studentPerCourse[k['Vizsgatargy2']] = 0
            
for i,k in data.iterrows():
    if pd.notnull(k['Vizsgaztato1']):
        studentPerCourse[k['Vizsgatargy1']] += 1
        if pd.notnull(k['Vizsgatargy2']):
            studentPerCourse[k['Vizsgatargy2']] += 1
            
studentPerCourse
```

```python
teacherPerCourse = {}
coursePerTeacher = {}
testPerTeacher = {}
courses = []

for i,k in data.iterrows():
    if pd.notnull(k['Vizsgaztato1']):
        if pd.notnull(k['Vizsgatargy1']):
            teacherPerCourse[k['Vizsgatargy1']] = 0
        if pd.notnull(k['Vizsgatargy2']):
            teacherPerCourse[k['Vizsgatargy2']] = 0
```

```python
for t in teachers:
    val = 0
    courses.clear()
    for i, k in data.iterrows():
        if k['Vizsgaztato1'] == t:
            courses.append(k['Vizsgatargy1'])
            val += 1
        elif k['Vizsgaztato2'] == t:
                val += 1
        if k['Vizsgaztato2'] == t:
            courses.append(k['Vizsgatargy2'])
    coursePerTeacher[t] = len(courses)
    testPerTeacher[t] = val
    for c in courses:
        teacherPerCourse[c] += 1

testPerTeacher
```

```python
studentPerTeacher = {}

for t in teachers:
    courses.clear()
    val = 0
    studentPerTeacher[t] = 0.0
    for i,k in data.iterrows():
        if k['Vizsgaztato1'] == t:
            courses.append(k['Vizsgatargy1'])
            val += 1
        elif k['Vizsgaztato2'] == t:
            val += 1
        if k['Vizsgaztato2'] == t:
            courses.append(k['Vizsgatargy2'])
    for c in courses:
        studentPerTeacher[t] += (studentPerCourse[c] / teacherPerCourse[c])

studentPerTeacher     
```
